```python
!pip install google-play-scraper

from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
pd.options.mode.chained_assignment = None
import csv
```

Scraping data dari review aplikasi (game) CSR Racing 2

```python
scrapreview = reviews_all(
    'com.naturalmotion.customstreetracer2',
    lang='en',
    country='us',
    sort=Sort.MOST_RELEVANT,
    count=15000
)
```

```python
with open('ulasan_CSR2.csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Review'])
  for review in scrapreview:
    writer.writerow([review['content']])
```

```python
df_hasil = pd.DataFrame(scrapreview)
df_hasil.to_csv('ulasan_CSR2.csv', index=False)
print(df_hasil.shape)

# Sepertinya limit 15000 di parameter sebelumnya tidak terbaca
# sehingga data yang diambil sebanyak 130500.
```
